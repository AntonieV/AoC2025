```python
### https://adventofcode.com/2025

import numpy as np
from scipy.spatial import ConvexHull
from matplotlib.path import Path


DAY = 9


def read_file(file_name):
    with open(file_name, 'r') as f_in:
        _data = list(filter(None, [line.strip().split(",") for line in f_in.readlines()]))
        return [(int(row), int(col)) for row, col in _data]
    

# def get_euclid_dist(p1, p2):
#     p1 = np.array(p1)
#     p2 = np.array(p2)
#     return round(float(np.sqrt(np.sum((p1 - p2)**2))), 4)


# def get_tiles_dists(_data):
#     points = []
#     for p1 in _data:
#         for p2 in _data:
#             if p1 != p2:
#                 points.append((get_euclid_dist(p1, p2), (p1, p2)))
#     return sorted(points, reverse=True)


def get_rectangle(p1, p2):
    (x1,y1), (x2, y2) = p1, p2
    l, w = abs(x1 - x2) + 1, abs(y1 - y2) + 1
    area =  w * l
    return area


def check_rectangle_in_hull(p1, p2, hull):
    (x1,y1), (x2, y2) = p1, p2
    start = (min(x1, x2), min(y1, y2))
    end = (max(x1, x2), max(y1, y2))
    for i_x in range(start[0], end[0] + 1):
        for i_y in range(start[1], end[1] + 1):
            if not hull.contains_point((i_x, i_y)):
                return False
    return True

    
def get_solution_1(points):
    rectangles = []
    for p1 in points:
        for p2 in points:
            rectangles.append(get_rectangle(p1, p2))
    return max(rectangles)
            

    
def get_solution_2(points):
    points = np.array(points)
    rectangles = []
#     print(points)
    hull = ConvexHull( points )
    hull_path = Path( points[hull.vertices] )
    for x1, y1 in points:
        print(x1, y1)
        for x2, y2 in points:
#             if x1 != x2 or y1 != y2:
            is_rect_in_hull = check_rectangle_in_hull((x1, y1), (x2, y2), hull_path)
#                 print(rect_points)
            if is_rect_in_hull:
                rectangles.append(get_rectangle((x1, y1), (x2, y2)))
    return max(rectangles)



file = f"../data/test_day_{DAY}.txt"
file = f"../data/day_{DAY}.txt"


data = read_file(file)

solution_1 = get_solution_1(data)
# solution_2 = get_solution_2(data)  # ToDo: does not work yet, runs forever
solution_2 = None

print(f"Day {DAY} - Solution 1:", solution_1)  # 4771532800
print(f"Day {DAY} - Solution 2:", solution_2)
```
