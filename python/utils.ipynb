````python
def get_graph(edges):
    g = dict()
    vertices = []
    for v1, v2 in edges:
        v1, v2 = str(v1), str(v2)
        vertices.append(v1)
        vertices.append(v2)
        if v1 not in g:
            g[v1] = []
        if v2 not in g[v1]:
            g[v1].append(v2)
    vertices = list(set(vertices))
    for v in vertices:
        if v not in g:
            g[v] = []
    return g, list(set(vertices))

# DFS
def has_path(start, end, visited, graph):
    if start == end:
        return True    
    if start not in visited:
        visited.append(start)
        for neighbour in graph[start]:
            result = has_path(neighbour, end, visited, graph)
            if result:
                return True
    return False


example = [
    (47, 53),
    (97, 13),
    (97, 61),
    (97, 47),
    (75, 29),
    (61, 13),
    (75, 53),
    (29, 13),
    (97, 29),
    (53, 29),
    (61, 53),
    (97, 53),
    (61, 29),
    (47, 13),
    (75, 47),
    (97, 75),
    (47, 61),
    (75, 61),
    (47, 29),
    (75, 13),
    (53, 13)    
]

paths = [
    ['75', '47', '61', '53', '29'],
    ['97', '61', '53', '29', '13'],
    ['75', '29', '13'],
    ['75', '97', '47', '61', '53'],
    ['61', '13', '29'],
    ['97', '13', '75', '29', '47']
]

graph, nodes = get_graph(example)

print(graph)

visits = []


sum_middle_valid = 0
for path in paths:
    has_path_all_nodes = []
    for idx, node in enumerate(path):
        if idx < len(path) - 1:
            v1, v2 = node, path[idx + 1]
            has_path_all_nodes.append(has_path(v1, v2, visits, graph))
    if all(has_path_all_nodes):
        sum_middle_valid += int(path[len(path)//2])
print(sum_middle_valid)
````
